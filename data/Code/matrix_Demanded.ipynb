```python
import pandas as pd

# 1) Read the original stations Ã— 365 matrix
#    The first column is the station ID index written by pandas
matrix_path = "410_data/matrix_14stations_365.csv"
matrix = pd.read_csv(matrix_path, index_col=0)

# 2') Compute row means (mean over all 365 days for each station)
row_means = matrix.mean(axis=1)

# 3') Subtract row means to get a demeaned matrix by station
#     Each row will now have mean ~ 0 across days
matrix_demeaned = matrix.sub(row_means, axis=0)


# 4) Save the demeaned matrix
demeaned_path = "410_data/matrix_14stations_365_demeaned_col.csv"
matrix_demeaned.to_csv(demeaned_path)

print("Saved demeaned (by column) matrix to:", demeaned_path)
print("Original shape:", matrix.shape, " Demeaned shape:", matrix_demeaned.shape)
```
