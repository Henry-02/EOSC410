```python
import pandas as pd

demeaned_path = "410_data/matrix_14stations_365_demeaned_col.csv"
X_demeaned = pd.read_csv(demeaned_path, index_col=0)

# Standard deviation of the original data by column
orig_path = "410_data/matrix_14stations_365.csv"
X_orig = pd.read_csv(orig_path, index_col=0)
col_stds = X_orig.std(axis=0, ddof=1)

Z = X_demeaned / col_stds
Z.to_csv("410_data/matrix_14stations_365_zscore_col.csv",)
```
